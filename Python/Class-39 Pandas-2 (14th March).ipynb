### How to handle the null values present in the DataFrame

#### How to find the null values in pandas 

Syntax: 
    
    isna() or isnull()

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
np.isnan(df.Age).sum()
```

```python
df.Age.isna().sum()
```

#### Fill the null values 

    Syntax:
        
        df.column_name.fillna(value)
        
- It does not make the changes in the existing file 

```python
df.Age.isna().sum()
```

```python
df.Age.mean()
```

```python
df.Age.fillna(df.Age.mean())
```

```python
df.Age.isna().sum()
```

```python
df.Age.fillna(df.Age.mean(), inplace = True)
```

#### How to deal with the Null values 

- When the missing values is less than 5% --> drop the rows 
- When the missing values is between 5% - 30% ---> impute the rows with either a mean or median with the column is numercial or mode when the column is categorical 
- When the missing values are greater than 30% ---> drop the column
