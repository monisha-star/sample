### Object Oriented Approach 

- This is also known as figure or axes approach
- In this appraoch, we can increase or decrease the size of the whole figure or a particular axes 

```python
plt.figure()
```

#### figure_object.add_axes([left_boundary, bottom_boundary, width, height])

```python
f = plt.figure()
f.add_axes([.1, .1, .4, .3])
f.add_axes([.7, .1, .5, .5])
f.add_axes([.7, .7, .8, .5])
f.add_axes([.2, .6, .5, .7])
```

### How to give title and the axes name in the object oriented appraoch 

Syntax:

        set_title('title')
        set_xlabel('Axis name')
        set_ylabel('Axis name')

```python
arr = np.arange(11)
arr2 = arr**2
```

```python
arr
```

```python
arr2
```

```python
f = plt.figure()
a = f.add_axes([.1, .1,.5, .5])
b = f.add_axes([.7, .1, .5, .5])

a.plot(arr, arr2, color = 'r', marker = 'o')
a.set_title('Plot axis - 1')
a.set_xlabel('arr')
a.set_ylabel('arr2')

b.plot(arr2, arr, color = 'g', marker = 'o')
a.set_title('Plot axis - 2')
a.set_xlabel('arr2')
a.set_ylabel('arr')
```

### How to increase the size of the figure

Syntax:

    plt.figure(figsize = (width, height), dpi = int)

    where dpi----> dots per inches 

```python
f = plt.figure(figsize = (10,10), dpi = 100)
a = f.add_axes([.1, .1,.5, .5])
b = f.add_axes([.7, .1, .5, .5])

a.plot(arr, arr2, color = 'r', marker = 'o')
a.set_title('Plot axis - 1')
a.set_xlabel('arr')
a.set_ylabel('arr2')

b.plot(arr2, arr, color = 'g', marker = 'o')
a.set_title('Plot axis - 2')
a.set_xlabel('arr2')
a.set_ylabel('arr')
```

### How to plot the data in a particular axes 

```python
f, a = plt.subplots(1,2, figsize = (12, 6), dpi = 80)

a[0].plot(arr, arr2, label = 'Arr vs Arr2', color = 'g', marker= 'o', markerfacecolor = 'y', markersize = 8, markeredgecolor = 'r', markeredgewidth  =2 )
a[0].set_title('Axes-1')
a[0].set_xlabel('arr')
a[0].set_ylabel('arr2')
a[0].legend(loc = 10)

a[1].plot(arr2, arr, label = 'Arr vs Arr2', color = 'r', marker= 'o', markerfacecolor = 'y', markersize = 8, markeredgecolor = 'g', markeredgewidth  =2 )
a[1].set_title('Axes-2')
a[1].set_xlabel('arr2')
a[1].set_ylabel('arr')
a[1].legend(loc = 10)
```
