### Groupby 

- It is used to analyse two or more columns together
- It is used to apply the aggregation function such as sum, values_Counts, min,mean, max, etc

Syntax:

        df.groupby('column').column.operation

```python
# I want to groupby Sex and find out the value count of the People belonging to different genders and whether they survived or not 

df.groupby('Sex').Survived.value_counts()
```

#### Groupby using multiple columns 

```python
df.groupby(['Pclass', 'Sex']).Survived.value_counts()
```

```python
# I want to find out the mean age of all the passengers classified as Pclass and Sex

df.groupby(['Pclass', 'Sex']).Age.mean()
```

```python
# I want to find out the max age of all the passengers classified as Pclass and Sex

df.groupby(['Pclass', 'Sex']).Age.max()
```

```python
# I want to find out the min age of all the passengers classified as Pclass and Sex

df.groupby(['Pclass', 'Sex']).Age.min()
```

```python
# Calculate the average fare and average age of the male and female passenger travelling in pclass

df.groupby(['Sex', 'Pclass'])[['Fare', 'Age']].mean()
```

```python
# Calculate the max fare and max age of the male and female passenger travelling in pclass

df.groupby(['Sex', 'Pclass'])[['Fare', 'Age']].max()
```

### Merging of the DataFrame 

    Syntax:

        pd.merge(df_1, df_2, on = common_column, how = )

```python
emp = pd.read_csv(r"C:\Users\Mayank\Downloads\Practice Datasets-20240314T143849Z-001\Practice Datasets\emp.csv")
emp
```

```python
dept = pd.read_csv(r"C:\Users\Mayank\Downloads\Practice Datasets-20240314T143849Z-001\Practice Datasets\dept.csv")
dept
```

#### how = inner(default)

- It takes all the common observation from both the Dataframe 

```python
pd.merge(emp, dept, on = 'Depart_No', how = 'inner')
```

#### how = 'left'
- All the observation from the left dataframe 

```python
pd.merge(emp, dept, on = 'Depart_No', how = 'left')
```

#### how = 'Right'

- All the observations from the right table 

```python
emp
```

```python
dept
```

```python
pd.merge(emp, dept, on = 'Depart_No', how = 'right')
```

#### how = 'outer'

- All the observations from  both the dataframe will be taken 

```python
pd.merge(emp, dept, on = 'Depart_No', how = 'outer')
```

#### Merging can be done on Multiple columns 

```python
pd.merge(emp, dept, on = ['Depart_No', 'City'], how = 'outer')
```

```python
pd.merge(emp, dept, on = ['Depart_No', 'City'], how = 'inner')
```

#### apply()

- It converts the column into an iterator
- The function (named function or lambda function) that we will pass in the apply() will be called on each observation of the column one by one

      Syntax:

                df.column.apply(function)


```python
(lambda x: 1 if x == 'male' else 0) ('male')
```

```python
(lambda x: 1 if x == 'male' else 0) ('female')
```
