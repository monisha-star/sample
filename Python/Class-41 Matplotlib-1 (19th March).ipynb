## Matplotlib

- It is a Python library that is used to plot 

### Plotting 

- The graphical representation of the data
- Plotting is an easier method to understand the nature of the data 

### How to import Matplotlib

```python
import matplotlib.pyplot as plt
```

### pyplot

- It is a module in matplotlib that contains all the function related to plotting the data 

```python
import pandas as pd
import numpy as np
```

```python
x = np.arange(10)
y = x**2
y
```

```python
x
```

```python
# plot the graph representing an exponential terms

plt.plot(x, y)
```

```python
x = [10,20,30,40]
y = [8, 4, 3, 7]

plt.plot(x, y)
```

### There are two types of approaches that we use to plot the data using Matplotlib

- Functional approach
- Object Oriented approach 

#### Functional Approach 

- It is also known as the beginner's approach
- It is used to plot simple plot quickly 

#### plot()

```python
x = np.array([i for i in range(1,11)])
x
```

```python
y = x**2
y 
```

```python
plt.plot(x, y)

plt.show()
```

#### Giving the title and the name to the axis 

    Syntax:

        plt.title('title_name')
        plt.xlabel('X_axis Name')
        plt.ylabel('Y_axis Name')

```python
plt.plot(x,y)
plt.title('Plot-1', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.xlabel('Arrat-X', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.ylabel('Array-Y', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
          
```

### Change the line style, color, line size

line style ---> ls = '-', '--', '-.', ',', ':'

line width -> lw = int/ float

marker = 'o', 'v', '*', '<' or '>'

```python
plt.plot(x, y, color = 'r', ls = '-', lw = '3', marker = 'o', markerfacecolor = 'yellow', markersize = 15, markeredgewidth = 2, markeredgecolor = 'g')
plt.title('Plot-1', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.xlabel('Arrat-X', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.ylabel('Array-Y', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
```

```python
plt.plot(x, y, color = 'r', ls = '--', lw = '3', marker = '*', markerfacecolor = 'yellow', markersize = 7, markeredgewidth = 2, markeredgecolor = 'g')
plt.title('Plot-1', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.xlabel('Arrat-X', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.ylabel('Array-Y', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
```

### Giving legend 

- It describe what is what 

```python
plt.plot(x, y, color = 'r', ls = '--', lw = '3', marker = '*', label = 'X vs Y',markerfacecolor = 'yellow', markersize = 7, markeredgewidth = 2, markeredgecolor = 'g')
plt.title('Plot-1', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.xlabel('Arrat-X', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})
plt.ylabel('Array-Y', fontdict = {'font': 'monospace', 'color' : 'red', 'size':20})

plt.legend(loc = 0, fontsize = 20, labelcolor = 'c')
plt.show()
```
