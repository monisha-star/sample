### Seaborn 

- It is a Python library, that is used to visualise the data
- It is an extended version of Matplotlib

### How to import seaborn ? 

```python
import seaborn as sns
```

### Version 

```python
sns.__version__
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### sns.set()

- It is used to plot a matplotlib plot on the seaborn default

```python
arr = np.arange(15)
arr
```

```python
arr1 = arr+5
arr1
```

### Matplotlib Plot 

```python
plt.plot(arr, arr1)
plt.show()
```

```python
sns.set()
plt.plot(arr, arr1)
plt.show()
```

### sns.set_style()

- It is used to set the grid system and other parameters within the seaborn

  Syntax:

              sns.set_style('grid_style') {darkgrid, whitegrid, white, dark}

```python
sns.set_style('darkgrid', {'figure.facecolor':'white', 'axes.facecolor':'grey', 'axesedgecolor':'0', 
                           'grid.linestyle':":", 'grid.color':'0'})
plt.plot(arr, arr1)
plt.show()
```

```python
sns.set_style('whitegrid', {'figure.facecolor':'white', 'axes.facecolor':'white', 'axesedgecolor':'0', 
                           'grid.linestyle':":", 'grid.color':'0'})
plt.plot(arr, arr1)
plt.show()
```

```python
sns.axes_style()
```

### sns.set_context()

- It is used for scalling

  notbook(default), poster, paper, talk

```python
sns.set_context('poster')
plt.plot(arr, arr1)
plt.show()
```

```python
sns.set_context('notebook')
plt.plot(arr, arr1)
plt.show()
```

```python
sns.set_context('paper')
plt.plot(arr, arr1)
plt.show()
```

```python
sns.set_context('talk')
plt.plot(arr, arr1)
plt.show()
```

### sns.set_themes()

    Syntax:

            sns.set_theme(style =, context = , palette =, font_scale)

```python
sns.set_theme(style = 'darkgrid', context = 'paper', font = 'monospace', font_scale = 1, palette = 'bright')
```

```python
plt.plot(arr, arr1, color = 'r', ls = '--', lw = 3, marker = 'o', label = 'X vs Y', markerfacecolor = 'y', markeredgewidth = 2, markersize = 15, markeredgecolor = 'g')
plt.title('Plot-1', fontdict = {'font':'monospace', 'color':'red', 'size':20})
plt.xlabel('Arr', fontdict = {'font':'monospace', 'color':'red', 'size':20})
plt.ylabel('Arr1', fontdict = {'font':'monospace', 'color':'red', 'size':20})
```

### Joinplot

- It returns a multi-panel figure
- IT plot the bi-variate and uni-variate distribution on each axes

  Syntax:

          sns.joinplot(x = discrete, y = discrete)
          kind = 'scatter(default)', 'reg', 'hist', 'hex'

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = data)
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length',kind = 'reg', data = data)
plt.show()
```

### pairplot

- It is used to compare one or more columns with one or more

  Syntax:

          sns.pairplot(x_vars = [column1, column2, ....], y_vars = [column_1, column_2,....], data = )
          kind = 'scatter(default)', 'reg', 'hist' , 'kde', 'hex'

```python
sns.pairplot(x_vars = ['sepal_length', 'petal_length'], y_vars = ['sepal_width', 'petal_width'], data = data)
plt.show()
```

```python
sns.pairplot(x_vars = ['sepal_length', 'petal_length'], y_vars = ['sepal_width', 'petal_width'], data = data, kind = 'kde')
plt.show()
```

```python
sns.pairplot(data, hue = 'species', kind = 'scatter', diag_kind = 'kde')
plt.show()
```

```python
sns.pairplot(df, hue = 'survived', kind = 'scatter', diag_kind = 'kde')
plt.show()
```
